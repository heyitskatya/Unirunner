```python
import random
import time

def unicorn_art():
    print("""
               \
                \\
                 \\\\
                  >\\/7
              _.-(6'  \\
             (=___._/` \\
                  )  \\ |
                 /   / |
                /    > /
               j    < _\\
           _.-' :      ``.
           \\ r=._\\        `.
          <`\\\\_  \\         .`-.
           \\ r-7  `-. ._  ' .  `\\
            \\`,      `-.`7  7) )
             \\/         \\|  \\' / `-._
                        ||    .'
                         \\\\  (
                          >\\  >
                      ,.-' >.'
                     <.'_.''
                       <'
    """)

print("🌈 Ласкаво просимо до Магічного Калькулятора з Єдинорогом! 🦄")
print("Вибери рівень складності:")
print("1 - Легкий 🍼")
print("2 - Середній 🧠")
print("3 - Складний 🧙‍♀️")

level = input("Введи 1, 2 або 3: ")

# Задачі для різних рівнів
easy = [
    ("1 + 1", 2),
    ("2 - 3", -1),
    ("4 * 2", 8),
    ("5 + 0", 5)
]

medium = [
    ("4 / 2 * 3", 6.0),
    ("7 - 10 + 4", 1),
    ("3 ** 2", 9),
    ("10 / 4", 2.5)
]

hard = [
    ("4 ** 3", 64),
    ("100 * 1.1 ** 5", round(100 * 1.1 ** 5, 2)),
    ("(5 + 3) * (2 ** 2)", 32),
    ("(10 - 3) ** 2 / 7", round((7)**2 / 7, 2))
]

# Вибираємо приклади за рівнем
if level == "1":
    tasks = easy
    print("👶 Починаємо з легкого рівня!")
elif level == "2":
    tasks = medium
    print("🧠 Готуйся! Середній рівень активовано.")
elif level == "3":
    tasks = hard
    print("🧙‍♂️ Ого! Ти обрав складний рівень!")
else:
    print("⚠️ Невірний вибір. Гра завершена.")
    exit()

# Починається гра
correct_answers = 0
for i in range(3):
    question, answer = random.choice(tasks)
    print(f"\n🔢 Завдання {i+1}: {question}")
    user_input = input("Ваша відповідь: ")

    try:
        user_answer = float(user_input)
        if abs(user_answer - answer) < 0.01:
            print("✅ Правильно! Єдиноріг плескає копитцями 🦄")
            correct_answers += 1
        else:
            print(f"❌ Неправильно. Правильна відповідь: {answer}")
    except:
        print("⚠️ Це не схоже на число. 😅")

time.sleep(1)
print("\n🎉 Гру завершено!")

# Підсумок
if correct_answers == 3:
    print("🌟 Ти справжній маг калькулятора! Ось твій чарівний єдиноріг:")
    unicorn_art()
elif correct_answers == 2:
    print("👍 Добре! Єдиноріг пишається тобою 🦄")
elif correct_answers == 1:
    print("🙂 Трохи ще потренуватися — і буде магія!")
else:
    print("😅 Схоже, магія ще попереду. Спробуй ще раз!")
```
